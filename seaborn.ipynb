```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Line Plot

```python
# line plot in matplotlib
x = [1,2,3,4,5]
y = [5,10,7,20,3]
plt.plot(x,y)
plt.show()
```

```python
# line plot in seaborn
df = pd.DataFrame({"x":[1,2,3,4],"y":[10,5,20,15]})
sns.lineplot(x="x",y="y", data=df)
plt.show()
```

```python
# load online dataset using seaborn
df = sns.load_dataset("penguins")
# Complete Line Plot in Seaborn
sns.lineplot(x="bill_length_mm", y = "bill_depth_mm", data=df, hue="sex",style="sex",palette="Accent", markers=["o","<"],dashes=False,legend=False)
plt.grid(axis="y")
plt.title("Python")
plt.show()
```

Bar Plot

```python
# bar plot in matplotlib
plt.bar(df['island'],df['bill_depth_mm'],color="r")
```

```python
# bar plot in seaborn
# create grid in bar plot
sns.set(style='darkgrid')
order_1 = ["Dream","Biscoe","Torgersen"]
sns.barplot(x="island",y='bill_depth_mm', data=df, palette="Accent",hue='sex', order=order_1,hue_order=["Female","Male"],ci=100,orient="v",color="r",saturation=0.6,errcolor="r",errwidth=1,capsize=0.1,alpha=0.8)
plt.title("Bar Plot using seaborn")
plt.show()
```

Histogram

```python
# Histogram plot in matplotlib
plt.hist(df['bill_depth_mm'],color="g")
plt.show()
```

```python
# Histogram plot in Seaborn
# in seaborn we can create histogram using displot
sns.displot(df['flipper_length_mm'])
plt.show()
```

```python
sns.displot(df['flipper_length_mm'],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True,color='g')
plt.show()
```

Scatter Plot

```python
# Scatter Plot in matplolib
plt.scatter(df['bill_depth_mm'],df['bill_length_mm'])
plt.show()
```

```python
# Scatter plot in seaborn
df = sns.load_dataset('penguins').head(20)
sns.scatterplot(x='bill_depth_mm',y='bill_length_mm',data=df,hue='sex',style='sex',size='sex',sizes=(100,50),palette='gist_ncar',alpha=0.5)
plt.show()
```

Heatmap

```python
# Heatmap in Matplotlib
```

```python
# Heatmap in seaborn
data = np.linspace(1,10,20).reshape(4,5)
sns.heatmap(data,vmin=1,vmax=10,cmap='PuOr',annot=True,linewidths=4,linecolor='pink',cbar=False)
```

```python
df = sns.load_dataset("anagrams")
x = df.drop(columns={'attnr'},axis=1).head(10)
sns.heatmap(x,annot=True,cmap='PuOr')
```

Count plot

```python
# count plot in seaborn
df = sns.load_dataset('tips')
# Vertical count plot
sns.countplot(x = df['sex'],data=df,hue='time',palette='PuOr',saturation=0.1)
plt.show()
```

```python
# Horizontal count plot
sns.countplot(y = df['sex'],data=df,hue='time',palette='PuOr',saturation=1)
plt.show()
```

Violine Plot

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='PuOr',inner='point')
plt.show()
```

Pair Plot

```python
sns.pairplot(df,hue='sex')
plt.show()
```

```python
# pair plot for all column
sns.pairplot(df,hue='sex',hue_order=['Female','Male'],palette='BuGn')
plt.show()
```

```python
# pair plot using particular columns
sns.pairplot(df,vars=["total_bill","tip"],hue='sex',hue_order=['Female','Male'],palette='BuGn',kind='scatter',markers=['*','>']) #Scatter, kde, hist, reg
plt.show()
```

Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=df,hue='sex',linewidth=1,edgecolor='r',alpha=0.5)
plt.show()
```

```python
sns.stripplot(x=df['total_bill'])
plt.show()
```

Box Plot

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data='df',order=['Fri','Sun','Thur','Sat'],showmeans=True,meanprops={"marker":"+",'markeredgecolor':'b'},linewidth=3)
```

Cat Plot

```python
# Catplot ek esa plot ya function h jiska use krke hum kisi bhi type k plot ko creat kr skte h with kind parameter
# stripplot() - with kind='strip'
# swarmplot() - with kind='swarm'
# boxplot() - with kind='box'
# violinplot() - with kind='violin'
# boxenplot() - with kind='boxen'
# pointplot() - with kind='point'
# barplot() - with kind='bar'
# countplot() - with kind='count'

df = sns.load_dataset('tips').head(20)
sns.catplot(x='tip',y='size',data=df,hue='sex',palette='PuOr',kind='point')
plt.show()
```

Styling Plots in Seaborn

```python
# Seaborn Figure Styles 
# Removing Axes Spines 
# Scale and Context
```

```python
df = sns.load_dataset('tips')
plt.figure(figsize=(4,4))

sns.set_style('dark') # use for dark background of chart
sns.set_style('darkgrid') # use for darkgrid background of chart

sns.set_context('poster',font_scale=1)

sns.barplot(x='day',y='total_bill',data=df)
sns.despine() #use for removing for axes line
plt.show()
```

FacetGrid Plot

```python
# Create 2 Plots
fg = sns.FacetGrid(df,col='sex')

fg.map(plt.scatter,"total_bill",'tip')

plt.show()
```
